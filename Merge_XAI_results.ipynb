```python
import pandas as pd
```

```python
dice_results = pd.read_csv('./results/DiCE_result_heart.csv')
alibi_cfrl_results = pd.read_csv('./results/Alibi_CFRL_result_heart.csv')
alibi_proto_results = pd.read_csv('./results/Alibi_Proto_result_heart.csv')

# 按行拼接（axis=0），忽略原有行索引，重新编号
df_concat_heart = pd.concat([dice_results, alibi_cfrl_results, alibi_proto_results], axis=0, ignore_index=True)

# 导出拼接后的结果
df_concat_heart.to_csv('./results/all_results_heart.csv', index=False)
```
