```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


nodes = ["X3", "X4", "X5", "X6"]

# =========================
# 1. CRPS 
# =========================
plt.figure(figsize=(7,4))
x = np.arange(len(nodes))
width = 0.35
plt.bar(x - width/2, crps_causal_random, width, label="Causal-Random")
plt.bar(x + width/2, crps_random, width, label="Random")
plt.xticks(x, nodes)
plt.ylabel("CRPS")
plt.title("CRPS Comparison (X3X6)")
plt.legend()
plt.savefig("./results/CRPS_Comparison.png")
plt.show()

# =========================
# 2. R 
# =========================
plt.figure(figsize=(7,4))
plt.bar(x - width/2, r2_causal_random, width, label="Causal-Random")
plt.bar(x + width/2, r2_random, width, label="Random")
plt.xticks(x, nodes)
plt.ylabel("R")
plt.title("R Comparison (X3X6)")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.legend()
plt.savefig("./results/R2_Comparison.png")
plt.show()

# =========================
# 3. KL divergence
# =========================
kl_labels = ["X1", "X2", "Overall"]
x = np.arange(len(kl_labels))
plt.figure(figsize=(6,4))
plt.bar(x - width/2, kl_causal_random, width, label="Causal-Random")
plt.bar(x + width/2, kl_random, width, label="Random")
plt.xticks(x, kl_labels)
plt.ylabel("KL Divergence")
plt.title("KL Divergence Comparison")
plt.legend()
plt.savefig("./results/KL_Comparison.png")
plt.show()

# =========================
# 4. Invertibility 
# =========================
invert_df = pd.DataFrame({
    "Causal-Random": invertibility_causal_random,
    "Random": invertibility_random
}, index=nodes)

plt.figure(figsize=(4,3))
sns.heatmap(invert_df.replace({"":1, "":0}).astype(int),
            annot=invert_df, fmt="", cmap="RdYlGn", cbar=False)
plt.title("Invertibility Check (X3X6)")
plt.savefig("./results/Invertibility_Comparison.png")
plt.show()


#  violated LMCs & Better than
graph_df_melt = graph_df.melt(id_vars="Method", value_vars=["Violated LMCs", "Better Than (%)"])
plt.figure(figsize=(6,4))
sns.barplot(data=graph_df_melt, x="variable", y="value", hue="Method")
plt.title("Graph Falsification Summary")
plt.savefig("./results/Graph_Falsification_Comparison.png")
plt.show()
```
